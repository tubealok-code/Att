# Export Repository Files to ZIP

This notebook scans the current repository, writes all files to a ZIP archive, and verifies the resulting archive contents.

## Import Required Libraries

Import Python modules such as pathlib, zipfile, and os to handle file paths and create the zip archive.

```python
from pathlib import Path
import zipfile
import os

repo_root = Path('.')
output_zip = repo_root / 'att-export.zip'
```

## Scan Repository Files

Traverse the repository directory tree and collect all files that should be included in the archive.

```python
exclude_dirs = {'.git', '__pycache__', '.venv'}
exclude_files = {'att-export.zip'}
files_to_zip = []

for path in repo_root.rglob('*'):
    if path.is_dir():
        continue
    if any(part in exclude_dirs for part in path.parts):
        continue
    if path.name in exclude_files:
        continue
    files_to_zip.append(path)

len(files_to_zip)
```

## Create ZIP Archive

Initialize a zipfile.ZipFile object in write mode and prepare the destination zip file.

```python
with zipfile.ZipFile(output_zip, 'w', compression=zipfile.ZIP_DEFLATED) as archive:
    for file_path in files_to_zip:
        archive.write(file_path, arcname=file_path.relative_to(repo_root))

output_zip.exists()
```

## Verify ZIP Contents

List the contents of the created zip file and confirm that all expected files were included.

```python
with zipfile.ZipFile(output_zip, 'r') as archive:
    zip_contents = archive.namelist()

len(zip_contents), zip_contents[:30]
```

## Optional: Exclude Patterns or Hidden Files

Show how to skip unwanted files or directories, such as .git folders or temporary files, before zipping.

```python
def should_include(path: Path) -> bool:
    if path.is_dir():
        return False
    if any(part.startswith('.') and part != '.gitignore' for part in path.parts):
        return False
    if path.suffix in {'.pyc', '.pyo', '.tmp', '.log'}:
        return False
    return True

filtered_files = [path for path in repo_root.rglob('*') if path.is_file() and should_include(path)]
len(filtered_files)
```
